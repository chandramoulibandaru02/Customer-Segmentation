```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
X=data.iloc[:,[3,4]].values
```

```python
X
```

```python

wcc=[]
```

```python
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(X)
    wcc.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcc)
plt.title('This is for finding the k best value for the clustering')
plt.xlabel('K values')
plt.ylabel('wcc')
plt.show()
```

```python
# here the break points are 3,5 for best part of clustering we are taking 5 as k value
```

```python
model=KMeans(n_clusters=5,init='k-means++')
```

```python
model.fit(x)
```

```python
y=model.predict(x)
```

```python
y
```

```python
plt.figure(figsize=(10,10),dpi=100)
plt.scatter(X[y==0,0],X[y==0,1],s=50,c='green',label='Cluster 1')
plt.scatter(X[y==1,0],X[y==1,1],s=50,c='red',label='Cluster 2')
plt.scatter(X[y==2,0],X[y==2,1],s=50,c='blue',label='Cluster 3')
plt.scatter(X[y==3,0],X[y==3,1],s=50,c='violet',label='Cluster 4')
plt.scatter(X[y==4,0],X[y==4,1],s=50,c='black',label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='cyan',label='Centroid')
plt.title('Customer grp')
plt.xlabel('Annual income ')
plt.ylabel('Spending score')
plt.legend()
plt.show()
```
